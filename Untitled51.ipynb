```python
# since we have skwed data in continous variable we are going to to log trasformation
for feature in continous_features:
  data=df.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
# findig if we have outliers in the data
# Loop through each continuous feature
for feature in continous_features:
    data = df.copy()  # Make a copy of the DataFrame to avoid changing the original

    # If the feature has a zero value, skip log transformation (log(0) is undefined)
    if 0 in data[feature].unique():
        pass  # Do nothing for this feature
    else:
        # Apply log transformation to reduce skewness and make outliers more visible
        data[feature] = np.log(data[feature])

        # Create a boxplot to visualize outliers
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()  # Show the plot

```

```python
# handling rare categorical features
# we we removing fearures that are present less that 1% of the observation
categorical_features
```

```python
# For training set
label_maps = {}
for feature in categorical_features:
    labels_ordered = df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered = {k: i for i, k in enumerate(labels_ordered, 0)}
    df[feature] = df[feature].map(labels_ordered)
    label_maps[feature] = labels_ordered  # Save the map to apply to test
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import train_test_split
```

```python
train=pd.read_csv("/content/x_train.csv")
```

```python
train.head()
```

```python
x_train=train.drop(['Id','SalePrice'],axis=1)
Y_train=train[['SalePrice']]
```

```python
feature_sel=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel.fit(x_train,Y_train)
```

```python
feature_sel.get_support()
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = x_train.columns[(feature_sel.get_support())]

# let's print some stats
print('total features: {}'.format((x_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel.estimator_.coef_ == 0)))
```

```python
x_train[selected_feat]
```

```python
model=Lasso(alpha=0.005,random_state=0)
model.fit(x_train[selected_feat],Y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred=model.predict(x_train[selected_feat])
mse=mean_squared_error(Y_train,y_pred)
print(mse)
```
